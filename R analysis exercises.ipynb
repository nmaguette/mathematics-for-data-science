```r
set.seed(42)
Y <- runif(40, 45, 100) ; t(Y)
```

```r
y <- Y[Y>=50 & Y<=70] ; t(y)
```

### 3. find the indices of these elements in vector y

```r
indices <- which(Y %in% y); indices # or match(y,Y) but it only returns the first occurence
```

a. between 50 and 60 or between 70 and 80

```r
y1 <- Y[(Y>50 & Y<60) | (Y>70 & Y<80)] ; y1 
```

b. equal to 50, 60 or 70

```r
y2 <- Y[(Y==50 | Y==60 | Y==70)] ; y2
```

c. different from 50, 60 and 70

```r
y3 <- Y[(Y!=50 | Y!=60 | Y!=70)] ; t(y3)
```

# Exercise 4

### 1. create a function that calculates the sum of two integers

```r
sum1 <- function(x, y){
  if(typeof(x)=='double' & typeof(y)=='double'){
    return(x+y)}
  else {
    print('one or two of the variables do not have the rigth type')}
}
sum1(0,0); sum1(3,-4); sum1('car',1); sum1('car','buy');
```

### 2. create a function that calculates x²-x-1

```r
polynome <- function(x){
  if(typeof(x)=='double'){
    return(x^2-x-1)}
  else {
    print('not the rigth type of variable')}
}
polynome(5); polynome(-2); polynome(0); polynome('house')
```

### 3. same question but by default we take x=0

```r
polynome <- function(x=0){
  if(typeof(x)=='double'){
    return(x^2-x-1)}
  else {
    print('not the rigth type')}
}
polynome()
```

### 4. write an is.multi5 function that returns TRUE if x is an integer divisible by 5

```r
is.multi5 <- function(x){
  if(typeof(x)=='double' & (x %% 5)==0)
    return(TRUE)
  FALSE
}
is.multi5(25); is.multi5(42); is.multi5(0)
```
