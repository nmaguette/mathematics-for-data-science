```python
# Problem 2
def BubbleSort(nums):
  # time complexity = O(n²)
  ch = nums[0]
  #print("func", ch, len(nums), range(0,len(nums),-1), list(range(0,len(nums))))
  for i in range((len(nums)-1),0,-1) :
    #print("loop", i)
    for j in range(0,(len(nums)-1)):
      if nums[j]>nums[j+1] :
        #print("loop", i, j, nums)
        ch=nums[j]
        nums[j]=nums[j+1]
        nums[j+1]=ch
  return nums 

print(BubbleSort([10,14,16,1,2,3,6,8]))
print(BubbleSort([16,1,5,2,10,14,3,6,8]))
print(BubbleSort([1,2,3,6,8,10,14,16]))
print(BubbleSort(list(range(15,0,-1))))
```

```python
#Dmitry solution

#Generate data module: for given sorting alrogithm and given array Data, generate points to plot
def Generate_Data(num_it, N_big, step, func):
   # func - sorting algorighm
   # num_it  - number of iterations to average over
   # N_big   - upper limit of computations
   # step    - step-size to plot
   #creating array points_to_plot of points to plot the graph
   np.random.seed(42)
   Data = np.random.rand(num_it, N_big)
   temp1 = np.array(range(1,int(np.log10(N_big)/step)))*step
   temp2 = np.ones(len(temp1)) * 10
   temp3 = np.power(temp2, temp1)
   points_to_plot = np.unique(temp3.astype(int))
   el_time = np.zeros(len(points_to_plot))
   for i in range(len(points_to_plot)):
       for j in range(num_it):
           temp = Data[j,:points_to_plot[i]]   #do this, because sort is working locally and change our Data
           t0 = time.time()
           func(temp)
           t1 = time.time()
           el_time[i] += (t1 - t0)
       el_time[i]/= num_it
   return(points_to_plot, el_time)

(points_to_plot, el_time) = Generate_Data(100, 200, 0.1, BubbleSort)
plt.figure(figsize=[12.8, 9.6])
plt.plot(points_to_plot, el_time)
plt.xlabel('Number of points', fontsize=15)
plt.ylabel('Time to run of the sorting algorithm', fontsize=15)
plt.title('Sorting data visualisation', fontsize=15)
plt.show()
```
