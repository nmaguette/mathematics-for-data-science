# Exercise 1

A researcher wants to analyze the relationship between class size (measured by the student-teacher ratio) and the average test score. He gets a dataset of class_size and test_score.
### 1. Create the vectors cs (class_size) et ts (test_score) containing the observartions

```r
(cs <- c(23, 19, 30, 22, 23, 29, 35, 36, 33, 25))
(ts <- c(430, 430, 333, 410, 390, 377, 325, 310, 33, 25))
```

### 2. Draw a scatterplot of the results

```r
plot(cs,ts)
```

### 3. Compute the mean, the sample variance and the sample standard deviation of ts

```r
mean(ts)
var(ts)
sd(ts)
```

### 4. Compute the covariance and the correlation coefficient for ts and cs 

```r
cov(cs, ts)
cor(cs, ts)
```

### 5. Use lm() to estimate the regression model 

```r
(lmMod <- lm(cs ~ ts))
```

### 6. Obtain a statistical summary of the model 

```r
summary(lmMod)
```

### 7. Add the regression to the scatterplot 

```r
plot(ts,cs)
abline(lmMod)
```

### 8. Can we valide this model?
No we can't, because as we see the points are far away from the model. Moreover in the summary, we can see that the **ts coefficient** is not very good.

# Exercise 3

We will create a datbase from the following two vectors


```r
set.seed(16)
x <- 0:50
y <- 2.3 - (15.1*x) + (1.2*x^2) + rnorm(length(x),20,50) #rnorm is a random normal data generated as a noise for our model
```

### 1. Create a data frame BDD using the two vectors year and population

```r
BDD <- data.frame(year=x, population=y) ; head(BDD)
```

### 2. Draw a scatterplot of the results

```r
pl <- plot(BDD$year, BDD$population)
```

### 3. Fit a model using polynomial regression

```r
(polReg <- lm(population~year+I(year^2), data=BDD))
```

### 4. Get an overview of the data

```r
summary(polReg)
```

```r
# trying to remove intercept for the polynomial regression model
(polReg_without_intercept <- lm(population~-1+year+I(year^2), data=BDD))
summary(polReg_without_intercept)
```

### 5. Can we validate the model?
Yes, we can validate this model, because as we see in the summary, we have good coefficients for p values (close to 0) and the R value is very close to 1.

And when we remove the intercept from the polynomial regression model, we get even better results. The reason is that the features are highly correlated.
